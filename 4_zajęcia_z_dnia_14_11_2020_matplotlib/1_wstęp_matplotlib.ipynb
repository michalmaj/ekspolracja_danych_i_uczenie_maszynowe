```python
# conda activate naza
# conda install -c conda-forge matplotlib
# anaconda doc
import matplotlib.pylab as plt
import matplotlib.pyplot as plt
```

```python
# pozwala na wyświetlanie figur bezpośrednio w jupyterze
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
# dla skrptów należy dodać jeszcz polecenie: plt.show()
```

```python
plt.plot(x,y)
# dodawanie opisów do osi
plt.xlabel('oś x')
plt.ylabel('oś y')

# dodawanie tytułu
plt.title('Tytuł')
```

```python
# wiele plotów:
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# obiektowe podejście:
fig = plt.figure() # tworzy figurę

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # dodawanie osi
axes.plot(x,y)

# tworzenie opisów osi:
axes.set_xlabel('oś x')
axes.set_ylabel('oś y')

# dodawanie tytułu
axes.set_title('Tytuł')
```

```python
# wiele plotów
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Większy")

axes2.plot(y,x)
axes2.set_title("Mniejszy")
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
# dodawanie subplotów
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
# przykład
fig,axes = plt.subplots(nrows=1,ncols=2)

# axes stanowią obiekt zawirający figury
# można się więc po nim iterować:
for ax in axes:
    ax.plot(x,y)
```

```python
# można też mieć do nich dostęp poprzez indeksację:
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title("Pierwsza")

axes[1].plot(y,x)
axes[1].set_title("Druga")
```

```python
# wielkość figury oraz liczba punktów obrazu przypadająca na cal (DPI)
```

```python
# wielkość figury
fig = plt.figure(figsize=(8,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# wielkosć sublotów
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,4))

axes[0].plot(x,y)
axes[1].plot(y,x,'r')

plt.tight_layout()
```

```python
# zapisywanie figury:
fig.savefig('moj_obrazek.png',dpi=600)
```

```python
# dodawanie legendy:
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,'g*',label='x do kwadratu')
ax.plot(x,x**3,'r-',label='x do sześcianu')

# dodanie parametru "loc=" pomoże ustwaić lokalizację legendy
# doc: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html
ax.legend(loc=0)
```

```python
# wygląd plotu
```

```python
# kolor
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#3737FF') # można także zastosować RGB w hexie
```

```python
# szerokość linii:
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='blue',linewidth=10)
```

```python
# przezroczystość linii:
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='blue',linewidth=10,alpha=0.3)
```

```python
# styl linii
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='blue',linewidth=10,linestyle='--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='blue',linewidth=1,linestyle='-',marker='o',markersize=10)
```

```python
# dodanie ficzerów do markerów:
fig = plt.figure(figsize=(12,8))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='blue',linewidth=1,linestyle='-',marker='o',markersize=10, markerfacecolor="red",
       markeredgewidth=3,markeredgecolor='orange')
```

```python
fig = plt.figure(figsize=(12,8))

ax = fig.add_axes([0,0,1,1])

# linewidth można zapisać jako lw, a linestyle jako ls
ax.plot(x,y,color='blue',linewidth=2,linestyle='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
